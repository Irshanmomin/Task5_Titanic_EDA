```python
numeric_df = df.select_dtypes(include=['number'])
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap (Numeric Columns Only)")
plt.show()
```

Observation:
The correlation map shows that Pclass is negatively correlated with Fare and Survived, meaning passengers in higher classes (lower Pclass number) had better survival chances.
Fare has a positive correlation with Survived, while Age shows a weak relationship. Other variables have very low correlation. 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pair_df = df[['Survived', 'Age', 'Fare', 'Pclass', 'SibSp', 'Parch']]
```

```python
sns.pairplot(pair_df, hue='Survived', diag_kind='kde')
plt.suptitle("Pairplot – Relationships Among Key Variables", y=1.02)
plt.show()
```

Observation:
The pairplot shows that survivors mostly belong to lower passenger classes (Pclass 1 or 2) and tend to have higher fares.  
Age has no clear separation, but younger and middle-aged passengers appear slightly more among survivors.  
Fare and Pclass show visible distinction between survivors and non-survivors.

```python
import pandas as pd
df = pd.read_csv(r"D:\train.csv")
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Pclass', y='Fare', hue='Survived', data=df)
plt.title("Fare vs Passenger Class – Survival Comparison")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Age Distribution by Survival")
plt.show()
```

Final Summary of Insights:

- Females and higher-class passengers had the highest survival rates.  
- Higher ticket fares were positively linked with survival.  
- Most passengers in 3rd class and males had lower survival chances.  
- Age showed a weak relationship, but younger passengers survived slightly more.  
- Social class and economic factors played a major role in survival outcomes.
